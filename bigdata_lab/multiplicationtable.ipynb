```python
lst = [3,4,7,2,9]
minn = maxx = lst[0]
for num in lst:
    if num > maxx:
        maxx = num
    elif num < minn:
        minn = num
print(f"maximum value is {maxx}")
print(f"minimum value is {minn}")
    
```
